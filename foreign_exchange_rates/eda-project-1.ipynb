````python
#Firstly I imported all libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
#Then I read csv file using pandas
pd_data=pd.read_csv("foreign_exchange_rates.csv")
pd_data.head()
````

````python
#I think, Unnamed: 0 column is not important that is why I deleted it
pd_data.drop(columns=["Unnamed: 0"], inplace=True)
pd_data.tail()
````

````python
pd_data.describe()
````

````python
columns=pd_data.columns.tolist()
print(columns)
````

````python
pd_data.info()
````

````python
#Replace ND to NaN
null=pd_data=="ND"
pd_data[null]=np.NaN
````

````python
#Drop NaN values
pd_data.dropna(inplace=True)
pd_data.index=np.arange(len(pd_data))

#Change dtypes
pd_data[pd_data.columns[1:]]=pd_data[pd_data.columns[1:]].astype('float')
````

````python
# Add new columns as year, month
from datetime import datetime
pd_data['Time Serie'] = pd.to_datetime(pd_data['Time Serie'])
pd_data["year"]=pd_data["Time Serie"].dt.year
pd_data["month"]=pd_data["Time Serie"].dt.month
pd_data["day"]=pd_data["Time Serie"].dt.day
pd_data["weekday"]=pd_data["Time Serie"].dt.weekday
pd_data['month_year'] = pd_data["Time Serie"].dt.to_period('M')
pd_data
````

````python

month=['Jan', 'Feb', 'March', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
for i in range(1,13):
    bool_month=pd_data['month']==i
    pd_data['month'][bool_month]=month[i-1]
````

````python
pd_data
````

````python
pd_data.info()
````

````python
# Dataset is ready to visual
pd_data.describe()
````

````python
# Group by year and plot 
grpb_year=pd_data.groupby('year').mean()
plt.rcParams["figure.figsize"]=(20,20)
fig=plt.figure()
axs=[]
for i in range(22):
    axs.append(fig.add_subplot(5, 5, i+1))
    axs[i].plot(grpb_year.index, grpb_year[grpb_year.columns[i]])
    axs[i].set_ylabel(grpb_year.columns[i])
    axs[i].set_xlabel('year')
    axs[i].set_xticks(np.arange(2000, 2020, 2));
    plt.setp(axs[i].get_xticklabels(), rotation=90, horizontalalignment='right')
    plt.tight_layout()
````

````python
# Group by month and plot 
grpb_month=pd_data.groupby('month').mean()
month=['Jan', 'Feb', 'March', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.rcParams["figure.figsize"]=(20,20)
fig=plt.figure()
axs=[]
for i in range(22):
    axs.append(fig.add_subplot(5, 5, i+1))
    axs[i].plot(month, grpb_month[grpb_month.columns[i]])
    axs[i].set_ylabel(grpb_month.columns[i])
    axs[i].set_xlabel('month')
    axs[i].set_xticks(month);
    plt.setp(axs[i].get_xticklabels(), rotation=90, horizontalalignment='right')
    plt.tight_layout()
````

````python
# Group by year and month and plot 
grpb_year_month=pd_data.groupby('month_year').mean()
x=grpb_year_month.index.astype(str)
plt.rcParams["figure.figsize"]=(20,20)
fig=plt.figure()
axs=[]
for i in range(22):
    axs.append(fig.add_subplot(5, 5, i+1))
    axs[i].plot(x, grpb_year_month[grpb_year_month.columns[i]])
    axs[i].set_ylabel(grpb_year_month.columns[i])
    axs[i].set_xticks(list(range(1, 240, 12)));
    axs[i].set_xticklabels([x for x in range(2000, 2020)])
    plt.setp(axs[i].get_xticklabels(), rotation=90, horizontalalignment='right')
    plt.tight_layout()
````
